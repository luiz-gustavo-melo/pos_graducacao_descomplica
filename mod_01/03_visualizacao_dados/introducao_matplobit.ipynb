```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Estudo com Dados aleat√≥rios')

plt.show
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['Ma√ß√£', 'Ma√ß√£ Verde', 'Laranja']  
counts = [10, 15, 7]
bar_labels = ['vermelho', 'verde', 'laranja']  # Labels for the bars  
bar_colors = ['tab:red', 'tab:green', 'tab:orange']  # Colors for the bars

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)


ax.legend(title='Cores das Frutas', loc='upper right')

plt.show()
```

```python
categorias = ['Categoria A', 'Categoria B', 'Categoria C']
valores = [20, 35, 30]

plt.bar(categorias, valores)
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Gr√°fico de Barras Simples')
plt.show()
```
