```python
import matplotlib.pyplot as plt
```

```python
# criando dados

x = [1, 2, 3, 4, 5]
y = [2, 5, 7, 9, 10]

# plotagem gráfico de linha
plt.bar(x,y)
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

#  edita as linhas de grade do gráfico
plt.grid(True, which='both', axis='y', color='gray', linestyle='-', linewidth=0.3, alpha=0.3)

# insere título do gráfico
plt.title("Gráfico com Grade Customizada")

# Mostrando o gráfico
plt.show()
```
